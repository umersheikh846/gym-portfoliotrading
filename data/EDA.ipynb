```python
import pandas as pd
import numpy as np

df = pd.read_csv("./prices-split-adjusted.csv")
df = df[df["symbol"].isin(["WMT","AAPL","MMM"])]
df_wmt = df[df["symbol"]=="WMT"]
df_aapl = df[df["symbol"]=="AAPL"]
```

```python
df_aapl = df_aapl.set_index("date")
df_wmt = df_wmt.set_index("date")
```

```python
# trading fee: 
# min 2 USD/deal
# 0.01 USD/stock

def get_trading_fee(price, trading_shares):
    fee = max(0.01*np.abs(trading_shares), 2)
    return fee

get_trading_fee(30,-300)
    
```

```python
portfolio ={"Cash":100000, "AAPL":0, "WMT":0}
portfolio
```

```python
date = df_aapl.index[0]
agent_1 = 100
agent_2 = 20
price_1 = df_aapl.at[date,"close"]
price_2 = df_wmt.at[date,"close"]
fee = get_trading_fee(price_1, agent_1) + get_trading_fee(price_2, agent_2)
portfolio["Cash"] = portfolio["Cash"] - fee - agent_1*price_1 - agent_2*price_2
portfolio["AAPL"] = portfolio["AAPL"] + agent_1
portfolio["WMT"] = portfolio["WMT"] + agent_2
portfolio
```
